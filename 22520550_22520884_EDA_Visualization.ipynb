# Outliers Visualization

```python
numeric_cols = matches.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(14, 6))
sns.boxplot(data=matches[numeric_cols])
plt.title("Boxplot for numeric columns")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

* Dựa trên biểu đồ về outliers ở các đặc trưng, phần lớn mỗi đặc trưng đều tồn tại các outliers, điều này có thể lý giải là do trong bóng đá sẽ thường xuất hiện những biến số nhất định, ví dụ như một đội tuyển có thể ghi được số bàn nhiều hơn bình thường do phong độ các cầu thủ đạt cực điểm,...

# Result Distribution of each team

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='result', hue='Home_team', data=matches, palette='Set2')
plt.title('Số trận thắng/hòa/thua của từng đội')
plt.xlabel('Kết quả trận đấu')
plt.ylabel('Số trận')
plt.legend(title='Đội tuyển', bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.tight_layout()
plt.show()
```

* Dựa trên phân bố kết quả của từng đội, có thể thấy rằng Manchester City có tổng số trận thắng là cao nhất, Crystal Palace có số trận thắng ít nhất, West Ham có số trận thua nhiều nhất,... Các yếu tố này có thể tác động đến dự đoán của mô hình khi các đội bóng có trình độ cao sẽ có khả năng chiến thắng cao hơn
