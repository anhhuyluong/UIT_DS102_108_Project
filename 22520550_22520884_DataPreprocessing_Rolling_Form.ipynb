# Gold Standard Processing

```python
def assign_season(df):
    """ThÃªm mÃ¹a giáº£i cho tá»«ng quan sÃ¡t Ä‘á»ƒ thá»±c hiá»‡n tÃ­nh Ä‘áº·c trÆ°ng Form"""
    df['Date'] = pd.to_datetime(df['Date'])

    conditions = [
        df['Date'] >= pd.Timestamp('2022-08-05'),
        df['Date'] >= pd.Timestamp('2021-08-14'),
        df['Date'] >= pd.Timestamp('2020-09-12'),
        df['Date'] >= pd.Timestamp('2019-08-09'),
        df['Date'] >= pd.Timestamp('2018-08-10'),
        df['Date'] >= pd.Timestamp('2017-08-12'),
    ]

    choices = [6, 5, 4, 3, 2, 1]

    df['Season'] = np.select(conditions, choices, default=7)

    return df
```

```python
matches = assign_season(matches)
matches
```

```python
def compute_form(df, gamma):
    """Táº¡o Ä‘áº·c trÆ°ng Form cho cÃ¡c quan sÃ¡t"""
    form_dict = {}
    teams = pd.unique(df[['hometeam_code', 'opponent_code']].values.ravel())

    df = df.copy()
    df['form_home'] = 1.0
    df['form_away'] = 1.0

    for season in df['Season'].unique():
        for team in teams:
            form_dict[(team, season)] = [1.0]

    for i, row in df.iterrows():
        team_a = row['hometeam_code']
        team_b = row['opponent_code']
        season = row['Season']
        target = row['target']

        form_a = form_dict[(team_a, season)][-1]
        form_b = form_dict[(team_b, season)][-1]

        if target == 1:
            new_form_a = form_a + gamma * form_b
            new_form_b = form_b - gamma * form_b
        elif target == 0:
            new_form_a = form_a - gamma * form_a
            new_form_b = form_b + gamma * form_a
        elif target == 2:
            new_form_a = form_a - gamma * (form_a - form_b)
            new_form_b = form_b - gamma * (form_b - form_a)
        else:
            continue

        form_dict[(team_a, season)].append(new_form_a)
        form_dict[(team_b, season)].append(new_form_b)

        df.at[i, 'form_home'] = new_form_a
        df.at[i, 'form_away'] = new_form_b

    return df
```

```python
def evaluate_gamma(df, gamma):
    df_with_form = compute_form(df, gamma)

    # Dá»± Ä‘oÃ¡n: náº¿u form_home > form_away => dá»± Ä‘oÃ¡n Home tháº¯ng
    y_pred = (df_with_form['form_home'] > df_with_form['form_away']).astype(int)
    y_true = df_with_form['target']

    accuracy = (y_pred == y_true).mean()
    return accuracy
```

```python
def search_best_gamma(df, gamma_values):
    best_gamma = None
    best_score = 0

    for gamma in gamma_values:
        score = evaluate_gamma(df, gamma)
        print(f"Gamma = {gamma:.2f}, Accuracy = {score:.4f}")
        if score > best_score:
            best_score = score
            best_gamma = gamma

    print(f"\nâœ… Best gamma: {best_gamma:.2f} with accuracy: {best_score:.4f}")
    return best_gamma
```

```python
gamma_range = np.arange(0.05, 1.05, 0.05)
best_gamma = search_best_gamma(matches, gamma_range)

# Táº¡o cá»™t Form tá»‘t nháº¥t
matches = compute_form(matches, best_gamma)
```

## Using Rolling Window to create new feature

```python
def create_rolling_feature(df, windows=3):
    """
    Sá»­ dá»¥ng Rolling Windows Ä‘á»ƒ tÃ­nh giÃ¡ trá»‹ trung bÃ¬nh cho cÃ¡c thÃ´ng sá»‘ á»Ÿ Ä‘á»™i Ä‘ang xÃ©t trong 3 tráº­n Ä‘áº¥u gáº§n nháº¥t diá»…n ra trÆ°á»›c Ä‘Ã³
    CÃ¡c Ä‘áº·c trÆ°ng Ä‘Æ°á»£c sá»­ dá»¥ng Ä‘á»ƒ tÃ­nh giÃ¡ trá»‹ trung bÃ¬nh gá»“m:
    - xg: Sá»‘ bÃ n tháº¯ng mong Ä‘á»£i
    - npxg: Sá»‘ bÃ n tháº¯ng mong Ä‘á»£i, khÃ´ng tÃ­nh cÃ¡c bÃ n tháº¯ng tá»« pháº¡t Ä‘á»n
    - shots: Sá»‘ cÃº sÃºt
    - tackles: Sá»‘ láº§n táº¯c bÃ³ng
    CÃ¡c Ä‘áº·c trÆ°ng Ä‘áº¡i diá»‡n cho cÃ¡c nhÃ³m táº¥n cÃ´ng, phÃ²ng thá»§
    """
    df = df.sort_values(by=['hometeam_code', 'Date']).copy()
    for stat in ['xg', 'npxg', 'shots', 'tackles']:
        df[f'{stat}_rolling'] = (
            df.groupby('hometeam_code')[stat]
            .shift(1)
            .rolling(windows)
            .mean()
        )
    return df
```

```python
def head_to_head_with_venue_both_directions(df, windows=3):
    """
    TÃ­nh toÃ¡n cÃ¡c Ä‘áº·c trÆ°ng Ä‘á»‘i Ä‘áº§u cÃ³ phÃ¢n biá»‡t sÃ¢n nhÃ  vÃ  sÃ¢n khÃ¡ch
    cho tá»«ng cáº·p Ä‘á»™i trong cáº£ hai chiá»u (Home vÃ  away) dá»±a trÃªn dá»¯ liá»‡u lá»‹ch sá»­

    CÃ¡c Ä‘áº·c trÆ°ng tÃ­nh gá»“m:
    - h2h_avg_result: trung bÃ¬nh káº¿t quáº£ tráº­n Ä‘áº¥u (tÃ­nh trÃªn target)
    - h2h_avg_goals: trung bÃ¬nh sá»‘ bÃ n tháº¯ng ghi Ä‘Æ°á»£c cá»§a Ä‘á»™i Ä‘ang xÃ©t
    - h2h_avg_goals_against: trung bÃ¬nh sá»‘ bÃ n thua pháº£i nháº­n cá»§a Ä‘á»™i Ä‘ang xÃ©t
    """
    df_copy = df.sort_values(by='Date').copy()

    # Khá»Ÿi táº¡o cÃ¡c Ä‘áº·c trÆ°ng
    for col in ['h2h_avg_result', 'h2h_avg_goals', 'h2h_avg_goals_against']:
        df_copy[f"{col}_{windows}"] = np.nan

    # Láº¥y táº¥t cáº£ cáº·p (home, away) khÃ¡c nhau
    unique_matchups = df_copy[['hometeam_code', 'opponent_code']].drop_duplicates()

    for _, row in unique_matchups.iterrows():
        team_a = row['hometeam_code']
        team_b = row['opponent_code']

        # --- Chiá»u A sÃ¢n nhÃ  ---
        mask_home = (df_copy['hometeam_code'] == team_a) & \
                    (df_copy['opponent_code'] == team_b) & \
                    (df_copy['venue_code'] == 1)

        df_home = df_copy[mask_home].sort_values(by='Date').copy()

        df_copy.loc[mask_home, f'h2h_avg_result_{windows}'] = (
            df_home['target'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_home, f'h2h_avg_goals_{windows}'] = (
            df_home['goals'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_home, f'h2h_avg_goals_against_{windows}'] = (
            df_home['goals_against'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )

        # --- Chiá»u A sÃ¢n khÃ¡ch ---
        mask_away = (df_copy['hometeam_code'] == team_a) & \
                    (df_copy['opponent_code'] == team_b) & \
                    (df_copy['venue_code'] == 0)

        df_away = df_copy[mask_away].sort_values(by='Date').copy()

        df_copy.loc[mask_away, f'h2h_avg_result_{windows}'] = (
            df_away['target'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_away, f'h2h_avg_goals_{windows}'] = (
            df_away['goals'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_away, f'h2h_avg_goals_against_{windows}'] = (
            df_away['goals_against'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )

    return df_copy
```
