# Silver Standard Processing

## Label Mapping Definition

```python
venue_mapping = {
    'Away': 0,
    'Home': 1
}

result_mapping = {
    'L': 0,
    'W': 1,
    'D': 2
}

team_mapping = {
    'Tottenham': 0,
    'Chelsea': 1,
    'Manchester Utd': 2,
    'Liverpool': 3,
    'Everton': 4,
    'Arsenal': 5,
    'Manchester City': 6,
    'Brighton': 7,
    'Newcastle Utd': 8,
    'West Ham': 9,
    'Crystal Palace': 10
}
```

## Encoding Object Columns

```python
matches['Date'] = pd.to_datetime(matches['Date'])
matches['venue_code'] = matches['venue'].map(venue_mapping)
matches['time'] = matches['start_time'].str.extract(r'(\d{2}:\d{2})')[0]
matches['time'] = matches['time'].apply(lambda x: int(x.split(':')[0]) + int(x.split(':')[1]) / 60)
matches['day_code'] = matches['Date'].dt.dayofweek
matches['hometeam_code'] = matches['Home_team'].map(team_mapping)
matches['opponent_code'] = matches['opponent'].map(team_mapping)
matches['target'] = matches['result'].map(result_mapping)
```

## Dropping unnecessary columns

```python
matches.drop(columns=['start_time', 'round', 'dayofweek', 'match_report', 'opponent', 'Home_team', 'result', 'goals_per_shot', 'venue'], inplace=True)
matches
```

## Convert Datatypes

```python
int_col = ['goals', 'shots', 'shots_on_target', 'shots_free_kicks', 'pens_made', 'pens_att', 'tackles', 'tackles_won', 'tackles_def_3rd', 'tackles_mid_3rd', 'tackles_att_3rd',
           'challenge_tackles', 'challenges', 'challenges_lost', 'blocks', 'blocked_shots', 'blocked_passes', 'interceptions', 'tackles_interceptions', 'clearances', 'errors']

for col in int_col:
    matches[col] = matches[col].astype('int64')
```

# Gold Standard Processing

## Using Rolling Window to create new feature

```python
def create_rolling_feature(df, windows=3):
    df = df.sort_values(by=['hometeam_code', 'Date']).copy()
    for stat in ['xg', 'npxg', 'shots', 'tackles']:
        df[f'{stat}_rolling'] = df.groupby('hometeam_code')[stat].shift(1).rolling(windows).mean()
    return df
```

```python
def head_to_head(df, windows=3):
    df = df.sort_values(by=['Date']).copy()

    df[f'h2h_avg_result_{windows}'] = (
        df.groupby(['hometeam_code', 'opponent_code'])['target']
        .shift(1).rolling(window=windows).mean()
    )

    df[f'h2h_avg_goals_{windows}'] = (
        df.groupby(['hometeam_code', 'opponent_code'])['goals']
        .shift(1).rolling(windows).mean()
    )

    df[f'h2h_avg_goals_against_{windows}'] = (
        df.groupby(['hometeam_code', 'opponent_code'])['goals_against']
        .shift(1).rolling(windows).mean()
    )

    return df
```

```python
def head_to_head_new_ver(df, windows=3):
    df_copy = df.sort_values(by=['Date']).copy() 
    df_copy[f'h2h_avg_result_{windows}'] = np.nan 
    df_copy[f'h2h_avg_goals_{windows}'] = np.nan
    df_copy[f'h2h_avg_goals_against_{windows}'] = np.nan

    # Láº¥y danh sÃ¡ch cÃ¡c cáº·p Ä‘á»™i duy nháº¥t
    unique_teams = df_copy[['hometeam_code', 'opponent_code']].drop_duplicates()

    for index, row in unique_teams.iterrows():
        home_team = row['hometeam_code']
        opponent_team = row['opponent_code']

        if home_team == opponent_team:
            continue 

        # Lá»c DataFrame cho cáº·p Ä‘á»™i hiá»‡n táº¡i
        mask = (df_copy['hometeam_code'] == home_team) & \
               (df_copy['opponent_code'] == opponent_team)

        # Ãp dá»¥ng logic tÃ­nh trung bÃ¬nh di Ä‘á»™ng cho pháº§n Ä‘Ã£ lá»c
        # LÆ°u Ã½: `.loc` Ä‘Æ°á»£c sá»­ dá»¥ng Ä‘á»ƒ cáº­p nháº­t cÃ¡c hÃ ng Ä‘Ã£ chá»n má»™t cÃ¡ch an toÃ n
        df_copy.loc[mask, f'h2h_avg_result_{windows}'] = (
            df_copy.loc[mask, 'target']
            .shift(1)
            .rolling(window=windows, min_periods=windows)
            .mean()
        )
        df_copy.loc[mask, f'h2h_avg_goals_{windows}'] = (
            df_copy.loc[mask, 'goals']
            .shift(1)
            .rolling(window=windows, min_periods=windows)
            .mean()
        )
        df_copy.loc[mask, f'h2h_avg_goals_against_{windows}'] = (
            df_copy.loc[mask, 'goals_against']
            .shift(1)
            .rolling(window=windows, min_periods=windows)
            .mean()
        )
    return df_copy
```

```python
def head_to_head_with_venue(df, windows=3):
    df_copy = df.sort_values(by='Date').copy()

    # Khá»Ÿi táº¡o cá»™t káº¿t quáº£
    for col in ['h2h_avg_result', 'h2h_avg_goals', 'h2h_avg_goals_against']:
        df_copy[f"{col}_{windows}"] = np.nan

    # Duyá»‡t táº¥t cáº£ cÃ¡c cáº·p chiá»u cÃ³ hÆ°á»›ng (team A lÃ  home, team B lÃ  away)
    unique_matchups = df_copy[['hometeam_code', 'opponent_code']].drop_duplicates()

    for _, row in unique_matchups.iterrows():
        home = row['hometeam_code']
        away = row['opponent_code']

        # Lá»c cÃ¡c dÃ²ng mÃ  home lÃ  chá»§ nhÃ  vÃ  away lÃ  Ä‘á»™i khÃ¡ch
        mask = (df_copy['hometeam_code'] == home) & (df_copy['opponent_code'] == away) & (df_copy['venue_code'] == 1)

        df_pair = df_copy[mask].sort_values(by='Date').copy()

        df_copy.loc[mask, f'h2h_avg_result_{windows}'] = (
            df_pair['target'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask, f'h2h_avg_goals_{windows}'] = (
            df_pair['goals'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask, f'h2h_avg_goals_against_{windows}'] = (
            df_pair['goals_against'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )

    return df_copy
```

```python
def head_to_head_with_venue_both_directions(df, windows=3):
    df_copy = df.sort_values(by='Date').copy()

    # Khá»Ÿi táº¡o cÃ¡c Ä‘áº·c trÆ°ng
    for col in ['h2h_avg_result', 'h2h_avg_goals', 'h2h_avg_goals_against']:
        df_copy[f"{col}_{windows}"] = np.nan

    # Láº¥y táº¥t cáº£ cáº·p (home, away) khÃ¡c nhau
    unique_matchups = df_copy[['hometeam_code', 'opponent_code']].drop_duplicates()

    for _, row in unique_matchups.iterrows():
        team_a = row['hometeam_code']
        team_b = row['opponent_code']

        # --- Chiá»u A sÃ¢n nhÃ  ---
        mask_home = (df_copy['hometeam_code'] == team_a) & \
                    (df_copy['opponent_code'] == team_b) & \
                    (df_copy['venue_code'] == 1)

        df_home = df_copy[mask_home].sort_values(by='Date').copy()

        df_copy.loc[mask_home, f'h2h_avg_result_{windows}'] = (
            df_home['target'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_home, f'h2h_avg_goals_{windows}'] = (
            df_home['goals'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_home, f'h2h_avg_goals_against_{windows}'] = (
            df_home['goals_against'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )

        # --- Chiá»u A sÃ¢n khÃ¡ch ---
        mask_away = (df_copy['hometeam_code'] == team_a) & \
                    (df_copy['opponent_code'] == team_b) & \
                    (df_copy['venue_code'] == 0)

        df_away = df_copy[mask_away].sort_values(by='Date').copy()

        df_copy.loc[mask_away, f'h2h_avg_result_{windows}'] = (
            df_away['target'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_away, f'h2h_avg_goals_{windows}'] = (
            df_away['goals'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )
        df_copy.loc[mask_away, f'h2h_avg_goals_against_{windows}'] = (
            df_away['goals_against'].shift(1).rolling(window=windows, min_periods=windows).mean()
        )

    return df_copy
```
